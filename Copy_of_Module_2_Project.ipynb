```python
from google.colab import drive
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#data loading
df=pd.read_csv("/content/Airbnb NYC 2019 (1).csv")
print(df)
#data description
print(df.head())

print(df.info())

print(df.describe())

#data cleaning

print(df.isnull().sum())
df.dropna(inplace=True)

#Data wrangling

print(df['room_type'].value_counts())

print(df.groupby('room_type')['price'].mean())

print(df['room_type'].value_counts()*df.groupby('room_type')['price'].mean())

print(df['neighbourhood_group'].value_counts())

print(df.groupby('neighbourhood_group')['price'].mean())

print(df['neighbourhood_group'].value_counts()*df.groupby('neighbourhood_group')['price'].mean())

#With respect to data wrangling we have correlated mean price to the neghbourhood.
#Also price to the room type is a metric to be considered.As both of these affect the booking frequency and price.



#data visualisation

sns.histplot(df['price'], bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()



sns.scatterplot(data=df, x='number_of_reviews', y='price', alpha=0.5)
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.title('Scatter Plot: Price vs. Number of Reviews')
plt.show()



sns.boxplot(data=df, x='neighbourhood_group', y='price')
plt.xlabel('Neighborhood Group')
plt.ylabel('Price')
plt.title('Price Distribution by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()


sns.scatterplot(data=df,x='room_type',y='price',alpha=0.5)
plt.xlabel('Room type')
plt.ylabel('Price')
plt.title("Price distribution by Room type")
plt.show()


plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

From the data at hand we can infer the following:

1.when considering overall sales and revenue bookings with entire home/apt have sold the most compared to private room which is second and shared room which have sold the least.
considering this properties having an entire apt/home to rent out are more likely to be booked.
2.Also, considering locality based booking, manhattan being the business district has generated most revenue followed by brooklyn. Hence in the future acquistion of more properties in manhattan and brooklyn will result in higher yearly revenue.
3.Also based on the sales numbers the pricing could be made more attractive to encourage bookings in higher value assets especially during holiday seasons and summer vacations.
